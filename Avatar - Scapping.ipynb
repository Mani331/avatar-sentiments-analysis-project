```python
import pandas as pd
import numpy as np
import requests
from requests import get
from bs4 import BeautifulSoup
```

```python
#first page to scrap
URL = requests.get("https://www.allocine.fr/film/fichefilm-61282/critiques/spectateurs/?page=1")
```

```python
soup = BeautifulSoup(URL.text, 'html.parser')
```

```python
soup.title
```

```python
for item1 in soup.find_all('span', {'class': 'stareval-note'}):
    print(item1.text)
for item2 in soup.find_all('div', {'class': 'content-txt review-card-content'}):
    print(item2.text)
```

```python
#checking if the number of pages is correct
urls = []
for i in range(1,719):
    urls.append("https://www.allocine.fr/film/fichefilm-61282/critiques/spectateurs/?page={}".format(i))

print(len(urls))
```

```python
rates = []
comments = []

for url in urls:
    URL = requests.get(url)
    soup = BeautifulSoup(URL.text, 'html.parser')
    print(url)
    
    for rate in soup.find_all('span', {'class':'stareval-note'}):
        rates.append(rate.text)
    
    for comment in soup.find_all('div', {'class': 'content-txt review-card-content'}):
        comments.append(comment.text)
    
    rates = rates[:-1]
```

```python
#checking if the size of our lists is the same
print(len(rates))
print(len(comments))
```

```python
#saving the data in a dataframe
Dict = {'rates':rates,'comments': comments}
df = pd.DataFrame(data = Dict)
```
